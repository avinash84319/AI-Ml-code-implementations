```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
df= pd.read_csv("mlr.csv")
```

```python
X=df[['age', 'experience']].values.astype(int)
Y=df.income
Y=Y/df.income.max()
```

```python
def cost(w,b):
    pred=np.dot(X,w)+b
    error=pred-Y
    return error
```

```python
def gradient(w,b,n,alpha):
    m=X.size
    error=cost(w,b)/m
    for i in range(n):
        w = w - np.dot(alpha*error,X)
        b = b - alpha*error.sum()
    return w,b
```

```python
w=np.array([0,0])
b=0
w,b=gradient(w,b,100,0.000012)
```

```python
def prediction(w,b,age,exp):
    return age*w[0] + exp*w[1] +b
```
