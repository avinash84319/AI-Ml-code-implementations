```python
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd
df=pd.read_csv('lr.csv')
```

```python
pl.scatter(df.diaBP,df.TenYearCHD)
```

```python
x=df.diaBP
y=df.TenYearCHD
x=x-x.mean()/(x.max()-x.min())
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def cost(w,b):
    m=x.count()
    loss=0
    for i in range(m):
        first= -y[i]*np.log(sigmoid(w*x[i]+b))
        second= -(1-y[i])*np.log(1-sigmoid(w*x[i]+b))
        loss+=(first+second)*x[i]
    cost=loss/m
    return cost
def cost2(w,b):
    m=x.count()
    loss=0
    for i in range(m):
        first= -y[i]*np.log(sigmoid(w*x[i]+b))
        second= -(1-y[i])*np.log(1-sigmoid(w*x[i]+b))
        loss+=first+second
    cost=loss/m
    return cost
```

```python
def gradientdes(w,b,alpha,n):
    for i in range(n):
        ws = w - alpha*cost(w,b)
        bs = b - alpha*cost2(w,b)
        w=ws
        b=bs

    return w,b
```

```python
w,b=gradientdes(0,0,0.00003,10)
print(w,b)
```

```python
pl.scatter(df.diaBP,sigmoid(x*w+b))
pl.scatter(df.diaBP,df.TenYearCHD)
```
