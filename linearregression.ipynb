```python

import pandas as pd
import numpy as n
import matplotlib.pyplot as m
df=pd.read_csv('gg.csv')
m.scatter(df.d,df.w,linewidth=1)
df.d.count()
points=df
def gradientdes(t1,t2,points,l):
    t1s=0
    t2s=0
    n=points.d.count()
    for i in range(n):
        x=points.d[i]
        y=points.w[i]
        t1s+=-(2/n)*x*(y-(x*t1+t2))
        t2s+=-(2*n)*(y-(x*t1+t2))
    
    t1=t1-t1s*l
    t2=t2-t2s*l
    
    return t1,t2
t1=0
t2=0
l=0.001
times=3000
for i in range(times):
    t1,t2=gradientdes(t1,t2,points,l)
print(t1,t2)
m.scatter(df.d,t1*df.d+t2,marker="d")
m.scatter(df.d,df.w,marker="v")
```
